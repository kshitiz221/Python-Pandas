```python
df.info()
```

```python
def last_four(num):
    return str(num)[-4:]
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
```

```python
df['total_bill'].mean()
```

```python
def review(price):
    if price < 10:
        return '$'
    elif price >= 10 and price <30:
        return '$$'
    else:
        return '$$$'
```

```python
df['review'] = df['total_bill'].apply(review)
```

```python
df.head()
```

```python
df[df['review'] == '$$$'].head()
```

```python
# Lambda Expressions
#lambda num: num**2
```

```python
df['total_bill'].apply(lambda num: num+2) # adds $2 to all the total_bill column
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Others'
```

```python
df['Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
```

```python
df.head()
```

**Using np.vectorize method**

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```
